# Garbage Collection

**Garbage Collection: is a process by which python automatically frees up memory by removing objects that are no longer in use**

*When we create variables, lists, dictionaries, etc. python stores them in memory(RAM). If you don't need them anymore, but they still stays in memory, your program will waste mempry.*

**Garbage Collector: python's gc module control the garbage collection process. It helps manage cyclic references-situations where two or more objects refers to each other**

*It is Python reference count overflow issue.*

*This happenes because: .Small integer like 0-256 are interned (Python keeps a a single shared copy of these values in memory for eficiency)*

*These small integer are used everywhere in Python-inside loops, constant, function calls, etc.*

**So their reference count become extremely large**

**The Real Reference Count is too large to fit in a 32-bit integer-it overflowed**

*This happenes beause: Python automatically caches(or interns)small integer objects in memory to save space and improve performance*

**Integer between -5 and 256 are pre-created and reused by Python.**
